```python
#Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the data
df_train = pd.read_csv('Train.csv')
df_test = pd.read_csv('Test.csv')
sample = pd.read_csv('Sample.csv')
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
sns.heatmap(df_train.isnull())
```

```python
edu_train_na = (df_train['education'].dropna().index)
prev_train_na = df_train['previous_year_rating'].dropna().index
edu_test_na = (df_test['education'].dropna().index)
prev_test_na = df_test['previous_year_rating'].dropna().index

edu_train_na = list(map(int,edu_train_na))
prev_train_na = list(map(int,prev_train_na))
edu_test_na = list(map(int,edu_test_na))
prev_test_na = list(map(int,prev_test_na))
```

```python
edu_train_drop = []
for i in range(54808):
    if i not in edu_train_na:
        edu_train_drop.append(i)

prev_train_drop = []
for i in range(54808):
    if i not in prev_train_na:
        prev_train_drop.append(i)

edu_test_drop = []
for i in range(23490):
    if i not in edu_test_na:
        edu_test_drop.append(i)

prev_test_drop = []
for i in range(23490):
    if i not in prev_test_na:
        prev_test_drop.append(i)
```

```python
df_edu_train_drop = df_train.iloc[edu_train_drop]
df_prev_train_drop = df_train.iloc[prev_train_drop]
df_edu_test_drop = df_test.iloc[edu_test_drop]
df_prev_test_drop = df_test.iloc[prev_test_drop]
```

```python
df_train_dropna = df_train.dropna()
```

```python
department = pd.get_dummies(df_train_dropna['department'])
gender = pd.get_dummies(df_train_dropna['gender'])
#age = pd.get_dummies(df_train_dropna['age'])
channel = pd.get_dummies(df_train_dropna['recruitment_channel'])
kpi = pd.get_dummies(df_train_dropna['KPIs_met >80%'])
awards = pd.get_dummies(df_train_dropna['awards_won?'])

df_train_dropna = pd.concat([department,gender,channel,kpi,awards,df_train_dropna['avg_training_score'],df_train_dropna['education']],axis=1)
```

```python
department = pd.get_dummies(df_edu_train_drop['department'])
gender = pd.get_dummies(df_edu_train_drop['gender'])
#age = pd.get_dummies(df_edu_train_drop['age'])
channel = pd.get_dummies(df_edu_train_drop['recruitment_channel'])
kpi = pd.get_dummies(df_edu_train_drop['KPIs_met >80%'])
awards = pd.get_dummies(df_edu_train_drop['awards_won?'])

df_edu_train_drop = pd.concat([department,gender,channel,kpi,awards,df_edu_train_drop['avg_training_score']],axis=1)
```

```python
department = pd.get_dummies(df_edu_test_drop['department'])
gender = pd.get_dummies(df_edu_test_drop['gender'])
#age = pd.get_dummies(df_edu_test_drop['age'])
channel = pd.get_dummies(df_edu_test_drop['recruitment_channel'])
kpi = pd.get_dummies(df_edu_test_drop['KPIs_met >80%'])
awards = pd.get_dummies(df_edu_test_drop['awards_won?'])

df_edu_test_drop = pd.concat([department,gender,channel,kpi,awards,df_edu_test_drop['avg_training_score']],axis=1)
```

```python
df_train['education'].value_counts()
```

```python
def func(string):
    if string == "Bachelor's":
        return 0
    if string == "Master's & above":
        return 1
    if string == "Below Secondary":
        return 2
```

```python
df_train_dropna['education'] = df_train_dropna['education'].apply(lambda x:func(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_train_dropna.drop('education',axis=1)
y = df_train_dropna['education']
```

```python
X_train,X_cv,y_train,y_cv = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import f1_score,accuracy_score,classification_report, confusion_matrix
```

```python
model = BernoulliNB()
model.fit(X_train,y_train)
pred = model.predict(X_cv)
```

```python
print(f1_score(pred,y_cv,average = 'macro'))
print()
print(accuracy_score(pred,y_cv))
print()
print(confusion_matrix(pred,y_cv))
print()
print(classification_report(pred,y_cv))
```

```python
pred = model.predict(df_edu_train_drop)
pred_test = model.predict(df_edu_test_drop)
```

```python
df_edu_train_drop = df_train.iloc[edu_train_drop]
df_edu_test_drop = df_test.iloc[edu_test_drop]
```

```python
df_edu_train_drop['education'] = pred
df_edu_test_drop['education'] = pred_test
```

```python
df_edu_train_drop.to_csv('df_edu_train_drop.csv')
df_edu_test_drop.to_csv('df_edu_test_drop.csv')
```
